```python
from faster_whisper import WhisperModel

# Choose model size: "small", "medium", or "large-v3"
model_size = "small"  # start with small; later you can try medium or large-v3

# Create model (will download on first run)
model = WhisperModel(model_size, device="cuda", compute_type="float16")
```

```python
from google.colab import files

uploaded = files.upload()  # this will open a file chooser
```

```python
audio_path = list(uploaded.keys())[0]
print("Using audio file:", audio_path)
```

```python
# language control:
#   language=None  -> auto-detect
#   "en" / "ar" / "tr" explicitly to force language

segments, info = model.transcribe(
    audio_path,
    beam_size=5,
    language=None  # set "en" or "ar" or "tr" if you want to force it
)

print("Detected language:", info.language)
print("Language probability:", info.language_probability)
print("-" * 40)

full_text = ""

for seg in segments:
    line = f"[{seg.start:.2f} → {seg.end:.2f}] {seg.text}"
    print(line)
    full_text += seg.text + " "

print("\n\nFull transcript:\n")
print(full_text.strip())
```

```python
output_txt = audio_path.rsplit(".", 1)[0] + "_transcript.txt"

with open(output_txt, "w", encoding="utf-8") as f:
    f.write(full_text.strip())

print("Saved transcript to:", output_txt)
```

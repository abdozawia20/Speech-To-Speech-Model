```python
import os
import subprocess
from IPython.display import Audio, display

# 1. Download the Piper Binary
# Using the stable 2023.11.14-2 release
piper_url = "https://github.com/rhasspy/piper/releases/download/2023.11.14-2/piper_linux_x86_64.tar.gz"
print(f"Downloading Piper from: {piper_url}")
!wget -q {piper_url}
!tar -xf piper_linux_x86_64.tar.gz

# 2. Download a Voice Model (en_US-lessac-medium)
print("Downloading voice model...")
!wget -q -O voice.onnx https://huggingface.co/rhasspy/piper-voices/resolve/main/en/en_US/lessac/medium/en_US-lessac-medium.onnx
!wget -q -O voice.onnx.json https://huggingface.co/rhasspy/piper-voices/resolve/main/en/en_US/lessac/medium/en_US-lessac-medium.onnx.json

# 3. Locate the binary (Safety Check)
piper_binary = "./piper/piper"
if not os.path.exists(piper_binary):
    print("Standard path not found. Searching...")
    # Fallback search
    result = !find . -name "piper" -type f -executable
    if result:
        piper_binary = result[0]
        print(f"Found binary at: {piper_binary}")
    else:
        raise FileNotFoundError("Could not find the 'piper' executable after extraction.")
```

```python
# 4. Run Inference
text_input = "This is Abdurrahman Zawia, and this file is part of the graduation project that is being conducted at Istanbul Bilgi University."
print(f"Generating audio for: '{text_input}'")

cmd = f'echo "{text_input}" | "{piper_binary}" --model voice.onnx --output_file output.wav'
subprocess.run(cmd, shell=True, check=True)

# 5. Play the result
if os.path.exists("output.wav") and os.path.getsize("output.wav") > 0:
    display(Audio("output.wav"))
else:
    print("Error: output.wav was not generated or is empty.")
```
